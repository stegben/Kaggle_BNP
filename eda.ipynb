```python
import pandas as pd
from IPython.display import display

df_train = pd.read_csv("train.csv")
df_test = pd.read_csv("test.csv")
```

```python
print(df_train["target"].value_counts())
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

for feat in df_train:
    if feat == "ID" or feat == "target":
        continue
    if df_train[feat].dtype == "object":
        print(df_train[feat].value_counts())
    else:
        df_train[feat].hist(bins=100)

```

```python

import matplotlib.pyplot as plt

for feat in df_train:
    print(feat)
    if feat == "ID" or feat == "target":
        continue
    if df_train[feat].dtype != "object":
        df_train[df_train["target"]==1][feat].hist(bins=100)
        df_train[df_train["target"]==0][feat].hist(bins=100)
        plt.show()
```

```python
from pandas.tools.plotting import parallel_coordinates
categorical_columns = df_train.columns[df_train.dtypes == 'object']

parallel_coordinates(df_train.drop(categorical_columns, axis=1), "target")
```
